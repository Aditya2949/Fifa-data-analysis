```python
# storing players salary and printing it
player_salary = fifa[['short_name','wage_eur']]
```

```python
# sort the players by its salrary on desending order
player_salary = player_salary.sort_values(by=['wage_eur'],ascending = False)
```

```python
# making bar plot for salary data
plt.figure(figsize=(7,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'][0:5]),color=("orange","yellow","pink","black","red"))
```
